```python
#Es diccionario, no string :/

def refactor(votes:str):
    lista=[]
    for couple in votes[1:-1].split(","):
        lista.append([couple.strip().split(":")[0][1:-1],couple.strip().split(":")[1]])
    
    for couple in lista:
        archivo[couple[0]] = couple[1]
    
    
#print(refactor("{'funny': 0, 'useful': 2, 'cool': 0}"))
```

```python
archivo["funny"] = archivo["stars"].apply( lambda x: list(x.values())[0] )
archivo["useful"] = archivo["votes"].apply( lambda x: list(x.values())[1] )
archivo["cool"] = archivo["votes"].apply( lambda x: list(x.values())[2] ) 
```

```python
def plot_por_tipo_y_fecha(dataframe, column: str, start_date: str, end_date:str, agg_fn = pd.DataFrame.mean,):
    
    #aquí filtro las reviews por fecha
    start_date=datetime.strptime(start_date,"%Y-%m-%d")
    end_date=datetime.strptime(end_date,"%Y-%m-%d")
    
    df = dataframe[(dataframe['fecha'] > start_date) & (dataframe['fecha'] <= end_date)]
    
    df_by_review = df.groupby([column,"fecha"])[["estrellas"]].aggregate(agg_fn)
    df_by_review.reset_index(inplace=True)
    df_by_review.set_index("fecha", inplace=True)

    df_by_review.plot(y = 'estrellas', legend=True, figsize=(24,8),title=column)
    
```

```python
def plot_por_tipo_y_fecha(dataframe, column: str, start_date: str, end_date:str, agg_fn = pd.DataFrame.mean,):
    
    #aquí filtro las reviews por fecha
    start_date=datetime.strptime(start_date,"%Y-%m-%d")
    end_date=datetime.strptime(end_date,"%Y-%m-%d")
    
    df = dataframe[(dataframe['fecha'] > start_date) & (dataframe['fecha'] <= end_date)]
    
    df_by_review=df[[column,"fecha","estrellas"]]

    df_by_review.plot(y = 'estrellas',x="fecha", legend=True, figsize=(24,8),title=column)
    
```

```python
for columna in ["divertido","útil","chido"]:
    plot_por_tipo_y_fecha(reviews,columna,"2012-01-01","2012-02-01")
```

```python
def create_boxplot_by_type(file_name:str, column: str, agg_fn= pd.DataFrame.count):
    df_complete = pd.read_csv(file_name)
    df_by_type = df_complete.groupby([column,"Fecha"])[["Sueldo Neto"]].aggregate(agg_fn)# .count()
    df_by_type.boxplot(by = column, figsize=(27,18))
    
def boxplot_por_tipo(dataframe, column: str, agg_fn = pd.DataFrame.mean,):
    df=dataframe
    df_by_review = df.groupby([column,"fecha"])[["estrellas"]].aggregate(agg_fn)
    df_by_review.reset_index(inplace=True)
    df_by_review.set_index("fecha", inplace=True)

    df_by_review.boxplot(by = column, figsize=(24,8))
```
